```
# Import libraries

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS 
```

Stopwords are collection of words that don't convey any meaning.

```
# Open the text file for reading 
text = open('JD.txt', 'r')

# Storing the content in a variable
text = text.read()
```

```
# Generate the wordcloud object, set the height and width, set the random_state parameter.

wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='blue', collocations=False, stopwords = STOPWORDS).generate(text)
```

```
# Plot the figure and set figure size
plt.figure(figsize=(40, 30))

# Display the wordcloud image
plt.imshow(wordcloud) 

# Exclude the axis 
plt.axis("off")
plt.show()
```
